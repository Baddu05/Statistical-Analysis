```python
#Question 2) An analyst at a department store wants to evaluate a recent credit card promotion. 
#To this end, 500 cardholders were randomly selected. Half received an ad promoting a reduced interest rate on 
#purchases made over the next three months,
#and half received a standard seasonal ad. Is the promotion effective to increase sales?
```

```python
#2Sample Ttest
```

```python
cust = pd.read_csv(r"C:\Users\badol\OneDrive\Documents\old laptop data\Business analytics course\8th july batch\Data Science And Python\data science foundation assignments\7. Basic Stats - Case Study-1\creditpromo.csv", engine = 'python')
cust
```

```python
std_promo = cust.dollars[(cust['insert'] == 'Standard')]
new_promo = cust.dollars[(cust['insert'] == 'New Promotion')]
```

```python
print (std_promo.head(3))
print (new_promo.head(3))
```

```python
print (std_promo.mean())
print (new_promo.mean())
```

```python
print (std_promo.std())
print (new_promo.std())
```

```python
stats.ttest_ind(a= std_promo,
                b= new_promo,
                equal_var= True)
```

```python
### Conclusions: The 2 sample t test was used to analyse the std_promo and new_promo levels measured to find the promotion
#effective to increase sales.

#stats negative sign indicates whatever value entered in a is less than value entered as b

#pvalue is less than 5% so we reject null hypothesis. Therefore, we can conclude that there is significant effect of promotion
#on sales.
```

```python
# Perform the ANOVA
stats.f_oneway(grp1, grp2, grp3, grp4, grp5, grp6)
```

```python
print (grp1.mean()); print (grp2.mean()); print (grp3.mean());print (grp4.mean()); print (grp5.mean()); print (grp6.mean())
```
